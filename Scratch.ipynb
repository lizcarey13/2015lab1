```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = y + np.random.normal(size=30) * .2
plt.plot(x, y, 'o-', label='A sine wave')
plt.plot(x, z, '-', label='Noisy sine')
plt.legend(loc= 'lower right')
plt.xlabel("X axis")
plt.ylabel("Y axis")
```
